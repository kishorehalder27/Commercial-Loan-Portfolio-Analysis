```python
# 1. Clean Interest Rate (which is inside df1)
df1['int_rate'] = df1['int_rate'].astype(str).str.replace('%', '', regex=False).astype(float)

# 2. Clean Revolving Utilization (which lives inside df2)
df2['revol_util'] = df2['revol_util'].astype(str).str.replace('%', '', regex=False).astype(float)

# 3. Verify both successfully changed to float numbers
print("df1 int_rate type:", df1['int_rate'].dtype)
print("df2 revol_util type:", df2['revol_util'].dtype)
```

```python
# Loop through both datasets to clean 'emp_length' simultaneously
for df_name, df in [('Finance 1', df1), ('Finance 2', df2)]:
    if 'emp_length' in df.columns:
        # Convert to string and strip characters
        df['emp_length'] = df['emp_length'].astype(str)
        df['emp_length'] = df['emp_length'].str.replace('years', '', regex=False)
        df['emp_length'] = df['emp_length'].str.replace('year', '', regex=False)
        df['emp_length'] = df['emp_length'].str.replace('+', '', regex=False)
        df['emp_length'] = df['emp_length'].str.replace('< 1', '0', regex=False)
        df['emp_length'] = df['emp_length'].str.strip()
        
        # Convert to clean integers and handle missing data
        df['emp_length'] = pd.to_numeric(df['emp_length'], errors='coerce').fillna(0).astype(int)
        print(f"✔️ {df_name} 'emp_length' successfully cleaned! Type: {df['emp_length'].dtype}")
```

```python
# 1. Clean issue_d inside Finance 1
if 'issue_d' in df1.columns:
    df1['issue_d'] = pd.to_datetime(df1['issue_d'], format='%b-%Y', errors='coerce')
    print("✔️ Finance 1 'issue_d' converted to true Date format!")

# 2. Clean last_pymnt_d inside Finance 2
if 'last_pymnt_d' in df2.columns:
    df2['last_pymnt_d'] = pd.to_datetime(df2['last_pymnt_d'], format='%b-%Y', errors='coerce')
    print("✔️ Finance 2 'last_pymnt_d' converted to true Date format!")
```

```python
print("--- Missing Values in Finance 1 ---")
# Display only columns that have at least one missing value
print(df1.isnull().sum()[df1.isnull().sum() > 0])

print("\n--- Missing Values in Finance 2 ---")
print(df2.isnull().sum()[df2.isnull().sum() > 0])
```

```python
# 1. Clean Finance 1 Missing Values
if 'title' in df1.columns:
    df1['title'] = df1['title'].fillna('Unknown')

# 2. Clean Finance 2 Missing Values
# If there are any missing rows in revolving utilization or dates, handle them safely
if 'revol_util' in df2.columns:
    df2['revol_util'] = df2['revol_util'].fillna(0.0)

if 'last_pymnt_d' in df2.columns:
    # For missing dates, we fill with a placeholder date or leave as NaT (Not a Time)
    # Leaving as NaT is standard for dates so it doesn't skew time calculations
    pass 

# 3. Double-check to ensure no major nulls remain in crucial columns
print("Finance 1 remaining nulls in title:", df1['title'].isnull().sum())
print("Finance 2 remaining nulls in revol_util:", df2['revol_util'].isnull().sum())
```

```python
# Check for duplicate IDs in both datasets
print("Duplicate IDs in Finance 1:", df1['id'].duplicated().sum())
print("Duplicate IDs in Finance 2:", df2['id'].duplicated().sum())
```
